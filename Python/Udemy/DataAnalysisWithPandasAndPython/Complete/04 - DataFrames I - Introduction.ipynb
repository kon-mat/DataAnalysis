## Methods and Attributes between Series and DataFrames
- A **DataFrame** is a 2-dimensional table consisting of rows and columns.

- Like with a **Series**, Pandas assigns an index position/label to each **DataFrame** row.
- The **DataFrame** and **Series** have common and exclusive methods/attributes.
- The `hasnans` attribute exists only a **Series**. The `columns` attribute exists only on a **DataFrame**.
- Some methods/attributes will return different types of data.
- The `info` method returns a summary of the pandas object.

## Differences between Shared Methods
- The `sum` method adds a **Series's** values.
- On a **DataFrame**, the `sum` method defaults to adding the values by traversing the index (row values).
- The `axis` parameter customizes the direction that we add across. Pass `"columns"` or `1` to add "across" the columns.

```python
revenue = pd.read_csv("revenue.csv", index_col="Date")
revenue
```

```python
pd.Series([1, 2, 3])
s.sum()
```

```python
# domyślnie metoda sum w ramce sumuje wartości w kolumnach
revenue.sum()
```

```python
revenue.sum(axis="index") # domyślna opcja
```

```python
# żeby utworzyć sumę w każdym wierszu ustawiamy parametr axis na "columns"
revenue.sum(axis="columns")
```

```python
# aby uzyskać sumę całości (wiersze+kolumny) używamy dwukrotnie funkcji sum
revenue.sum(axis="columns").sum()
```

- Pandas extracts a column from a **DataFrame** as a **Series**.
- The **Series** is a view, so changes to the **Series** *will* affect the **DataFrame**.
- Pandas will display a warning if you mutate the **Series**. Use the `copy` method to create a duplicate.

## Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
# żeby usunąć wiersze z brakującymi wartościami (NaN) możemy wykorzystać metodę dropna
# nie ma znaczenia czy brkaująca wartość (NaN) występuje w jednej kolumnie, czy całym wierszu
nba.dropna()
```

```python
# jeżeli chcemy usunąć wyłącznie wiersze, które nie mają wartość w każdej kolumnie (cały wiersz pusty),
# to możemy wykorzystać parametr how="all" dla metody dropna
# wtedy kolumny z częściowo brakującymi wartościami nie zostaną usunięte
nba.dropna(how="all")
```

```python
# jeżeli chcemy ograniczyć wyszukiwanie NaN do konkretnych kolumn, to możemy zdefiniwać parametr subset
nba.dropna(subset=["College"])  # w tym wypadku wiersz z wartościa NaN w kolumnie Salary nie zostanie usunięty
```

```python
nba.dropna(subset=["Salary"])
```

```python
# do subset możemy podawać więcej argumentów
nba.dropna(subset=["College", "Salary"])
```

## Fill in Missing Values with the fillna Method
- The `fillna` method replaces missing `NaN` values with its argument.
- The `fillna` method is available on both **DataFrames** and **Series**.
- An extracted **Series** is a view on the original **DataFrame**, but the `fillna` method returns a copy.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba
```

```python
# metoda fillna powoduje wypełnienie każdej NaN wartością podaną w argumencie
# nba tworzy nową kopię i w niej wypełnia NaN, oryginalny element pozostaje nienaruszony
nba.fillna(0)
```

```python
# metodę fillna możemy wykorzystać również na serii
nba["Salary"].fillna(0)
```

```python
# żeby zapisać zmiany w serii musimy wykorzystać widok serii
# w ten sposób nie nadpiszemy wartości Nan w nieodpowiednich kolumnach
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
# w tekstowym typie danych możemy nadpisać NaN za pomocą ciągu znaków
nba["College"] = nba["College"].fillna(value="Unknown")
```

```python
nba
```

## The astype Method I
- The `astype` method converts a **Series's** values to a specified type.
- Pass in the specified type as either a string or the core Python data type.
- Pandas cannot convert `NaN` values to numeric types, so we need to eliminate/replace them before we perform the conversion.
- The `dtypes` attribute returns a **Series** with the **DataFrame's** columns and their types.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba
```

```python
# żeby przekonwertować wartości z danej kolumny na liczby całkowite, musimy pozbyć się z niej NaN
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
# dtypes informuje nas o aktualnych typach dla poszczególnych kolumn
nba.dtypes
```

```python
# żeby przekonwetować kolumnę na dany typ wykorzystujemy metodę astype
# argumentem, który musimy podać jest typ, na który konwertujemy (możemy go podać w "" lub bez)
# tym sposobem tworzymy kopię oryginalnego obiektu
nba["Salary"].astype("int")
nba["Salary"].astype(int)
```

```python
# standardowo możemy zmienić wartości w oryginalnym obiekcie
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba
```

```python
nba.dtypes
```

```python
nba["Weight"] = nba["Weight"].fillna(0)
nba["Weight"] = nba["Weight"].astype("int")
nba
```

## The astype Method II
- The `category` type is ideal for columns with a limited number of unique values.
- The `nunique` method will return a **Series** with the number of unique values in each column.
- With categories, pandas does not create a separate value in memory for each "cell". Rather, the cells point to a single copy for each unique value.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba.tail()
```

```python
# nunique zwraca ilość unikalnych rekordów w kolumnie
nba["Team"].nunique()
```

```python
# gdy wywołujemy metodę na ramcę, to otrzymujemy wartości dla każdej z kolumn
# ta metoda może nam wskazić najlepszą kolumnę do zdefiniowania kategorii (najmniejsza liczba unikalnych wartości)
nba.nunique()
```

```python
# bez zdefiniowanych kolumn nasze zużycie pamięci wynosi 36,9 KB
nba.info()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
# powyższą metodą zredukowaliśmy zużycie pasmięci do 33,2 KB
nba.info()
```

```python
nba["Team"] = nba["Team"].astype("category")
```

```python
# bez zdefiniowanych kolumn nasze zużycie pamięci wynosi 30,5 KB
nba.info()
```

```python
# konwersja ns string
nba["Team"] = nba["Team"].astype("str")
nba
```

## Sort a DataFrame with the sort_values Method I
- The `sort_values` method sorts a **DataFrame** by the values in one or more columns. The default sort is an ascending one (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- If sorting by a single column, pass a string with its name.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# metoda sort_values sortuje serię w kolejności rosnącej
nba["Name"].sort_values()
```

```python
# sort_values w przypadku ramki nie zadziała bez argumentów, ponieważ musimy zdefiniować przynajmniej jedną kolumnę do sortowania
# parametr by definiuje kolumny, na podstawie których chcemy sortować
nba.sort_values(by="Name")
```

```python
# możemy również sortować w kolejności descending
# domyślnie pandas umieszcza wartości NaN na końcu (niezależnie od kierunku sortowania)
nba.sort_values(by="Salary", ascending=False)
```

```python
# za pomocą parametru na_position możemy zdefiniować, czy wartości NaN mają pojawić się na początku/końcu
nba.sort_values(by="Salary", ascending=False, na_position="first")
```

## Sort a DataFrame with the sort_values Method II
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# możemy również sortować po kilku kolumnach, ale podajemy w parametrze listę kolumn
# jest to sortowanie sekwencyjne i wpisujemy kolejne kolumny zgodnie z sekwencją sortowania
# główne sortowanie jest na podstawie kolumny "Team", a następnie kolejne na podstawie kolumny "Name"
nba.sort_values(by=["Team", "Name"])
```

```python
# jeżeli chcemy sortować różne kolumny w różnych kolejnościach, to w parametrze ascending podajemy listę wartości True/False
nba.sort_values(by=["Team", "Name"], ascending=[True, False])
```

```python
# sortowanie jest wykonywane na kopii oryginalnego obiektu
nba.sort_values(by=["Position", "Salary"], ascending=[True, False])
```

```python
# jeżeli chcemy sortowanie zapisać w oryginalnym obiekcie, to musimy użyć przypisania
nba = nba.sort_values(by=["Position", "Salary"], ascending=[True, False])
nba
```

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each **Series** value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba["Salary"]
```

```python
# jeżeli chcemy utworzyć ranking na podstawie zdefiniowanych wartości, to możemy wykorzystać metodę rank
# metoda rank zwróci nam kolumnę z miejscem pracownika w rankingu
nba["Salary"].rank()
```

```python
# jeżeli wykorzystamy parametr ascending=False, to ranking zostanie zdefiniowany w kolejności descending (od największych do najmniejszych)
# możemy zwrócić te wartości jako liczby całkowite
nba["Salary"].rank(ascending=False).astype("int")
```

```python
# ranking możemy dodać jako kolejna kolumnę
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
nba
```

```python
# jeżeli jakieś rekordy z rankingu będą mieć te same wartości, to powstaje remis i wartość rankingowe się powtarzają
nba.sort_values(by="Salary", ascending=False).head(10)
```
