## Differences between Shared Methods
- The `sum` method adds a **Series's** values.
- On a **DataFrame**, the `sum` method defaults to adding the values by traversing the index (row values).
- The `axis` parameter customizes the direction that we add across. Pass `"columns"` or `1` to add "across" the columns.

```python
revenue = pd.read_csv("revenue.csv", index_col="Date")
revenue
```

```python
pd.Series([1, 2, 3])
s.sum()
```

```python
# domyÅ›lnie metoda sum w ramce sumuje wartoÅ›ci w kolumnach
revenue.sum()
```

```python
revenue.sum(axis="index") # domyÅ›lna opcja
```

```python
# Å¼eby utworzyÄ‡ sumÄ™ w kaÅ¼dym wierszu ustawiamy parametr axis na "columns"
revenue.sum(axis="columns")
```

```python
# aby uzyskaÄ‡ sumÄ™ caÅ‚oÅ›ci (wiersze+kolumny) uÅ¼ywamy dwukrotnie funkcji sum
revenue.sum(axis="columns").sum()
```

## Drop Rows with Missing Values
- Pandas uses a `NaN` designation for cells that have a missing value.
- The `dropna` method deletes rows with missing values. Its default behavior is to remove a row if it has *any* missing values.
- Pass the `how` parameter an argument of "all" to delete rows where all the values are `NaN`.
- The `subset` parameters customizes/limits the columns that pandas will use to drop rows with missing values.

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
# Å¼eby usunÄ…Ä‡ wiersze z brakujÄ…cymi wartoÅ›ciami (NaN) moÅ¼emy wykorzystaÄ‡ metodÄ™ dropna
# nie ma znaczenia czy brkaujÄ…ca wartoÅ›Ä‡ (NaN) wystÄ™puje w jednej kolumnie, czy caÅ‚ym wierszu
nba.dropna()
```

```python
# jeÅ¼eli chcemy usunÄ…Ä‡ wyÅ‚Ä…cznie wiersze, ktÃ³re nie majÄ… wartoÅ›Ä‡ w kaÅ¼dej kolumnie (caÅ‚y wiersz pusty),
# to moÅ¼emy wykorzystaÄ‡ parametr how="all" dla metody dropna
# wtedy kolumny z czÄ™Å›ciowo brakujÄ…cymi wartoÅ›ciami nie zostanÄ… usuniÄ™te
nba.dropna(how="all")
```

```python
# jeÅ¼eli chcemy ograniczyÄ‡ wyszukiwanie NaN do konkretnych kolumn, to moÅ¼emy zdefiniwaÄ‡ parametr subset
nba.dropna(subset=["College"])  # w tym wypadku wiersz z wartoÅ›cia NaN w kolumnie Salary nie zostanie usuniÄ™ty
```

```python
nba.dropna(subset=["Salary"])
```

```python
# do subset moÅ¼emy podawaÄ‡ wiÄ™cej argumentÃ³w
nba.dropna(subset=["College", "Salary"])
```

## Fill in Missing Values with the fillna Method
- The `fillna` method replaces missing `NaN` values with its argument.
- The `fillna` method is available on both **DataFrames** and **Series**.
- An extracted **Series** is a view on the original **DataFrame**, but the `fillna` method returns a copy.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba
```

```python
# metoda fillna powoduje wypeÅ‚nienie kaÅ¼dej NaN wartoÅ›ciÄ… podanÄ… w argumencie
# nba tworzy nowÄ… kopiÄ™ i w niej wypeÅ‚nia NaN, oryginalny element pozostaje nienaruszony
nba.fillna(0)
```

```python
# metodÄ™ fillna moÅ¼emy wykorzystaÄ‡ rÃ³wnieÅ¼ na serii
nba["Salary"].fillna(0)
```

```python
# Å¼eby zapisaÄ‡ zmiany w serii musimy wykorzystaÄ‡ widok serii
# w ten sposÃ³b nie nadpiszemy wartoÅ›ci Nan w nieodpowiednich kolumnach
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
# w tekstowym typie danych moÅ¼emy nadpisaÄ‡ NaN za pomocÄ… ciÄ…gu znakÃ³w
nba["College"] = nba["College"].fillna(value="Unknown")
```

```python
nba
```

## The astype Method I
- The `astype` method converts a **Series's** values to a specified type.
- Pass in the specified type as either a string or the core Python data type.
- Pandas cannot convert `NaN` values to numeric types, so we need to eliminate/replace them before we perform the conversion.
- The `dtypes` attribute returns a **Series** with the **DataFrame's** columns and their types.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba
```

```python
# Å¼eby przekonwertowaÄ‡ wartoÅ›ci z danej kolumny na liczby caÅ‚kowite, musimy pozbyÄ‡ siÄ™ z niej NaN
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
# dtypes informuje nas o aktualnych typach dla poszczegÃ³lnych kolumn
nba.dtypes
```

```python
# Å¼eby przekonwetowaÄ‡ kolumnÄ™ na dany typ wykorzystujemy metodÄ™ astype
# argumentem, ktÃ³ry musimy podaÄ‡ jest typ, na ktÃ³ry konwertujemy (moÅ¼emy go podaÄ‡ w "" lub bez)
# tym sposobem tworzymy kopiÄ™ oryginalnego obiektu
nba["Salary"].astype("int")
nba["Salary"].astype(int)
```

```python
# standardowo moÅ¼emy zmieniÄ‡ wartoÅ›ci w oryginalnym obiekcie
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba
```

```python
nba.dtypes
```

```python
nba["Weight"] = nba["Weight"].fillna(0)
nba["Weight"] = nba["Weight"].astype("int")
nba
```

## The astype Method II
- The `category` type is ideal for columns with a limited number of unique values.
- The `nunique` method will return a **Series** with the number of unique values in each column.
- With categories, pandas does not create a separate value in memory for each "cell". Rather, the cells point to a single copy for each unique value.

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba.tail()
```

```python
# nunique zwraca iloÅ›Ä‡ unikalnych rekordÃ³w w kolumnie
nba["Team"].nunique()
```

```python
# gdy wywoÅ‚ujemy metodÄ™ na ramcÄ™, to otrzymujemy wartoÅ›ci dla kaÅ¼dej z kolumn
# ta metoda moÅ¼e nam wskaziÄ‡ najlepszÄ… kolumnÄ™ do zdefiniowania kategorii (najmniejsza liczba unikalnych wartoÅ›ci)
nba.nunique()
```

```python
# bez zdefiniowanych kolumn nasze zuÅ¼ycie pamiÄ™ci wynosi 36,9 KB
nba.info()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
# powyÅ¼szÄ… metodÄ… zredukowaliÅ›my zuÅ¼ycie pasmiÄ™ci do 33,2 KB
nba.info()
```

```python
nba["Team"] = nba["Team"].astype("category")
```

## Sort a DataFrame with the sort_values Method I
- The `sort_values` method sorts a **DataFrame** by the values in one or more columns. The default sort is an ascending one (alphabetical for strings).
- The first parameter (`by`) expects the column(s) to sort by.
- If sorting by a single column, pass a string with its name.
- The `ascending` parameter customizes the sort order.
- The `na_position` parameter customizes where pandas places `NaN` values.

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# metoda sort_values sortuje seriÄ™ w kolejnoÅ›ci rosnÄ…cej
nba["Name"].sort_values()
```

```python
# sort_values w przypadku ramki nie zadziaÅ‚a bez argumentÃ³w, poniewaÅ¼ musimy zdefiniowaÄ‡ przynajmniej jednÄ… kolumnÄ™ do sortowania
# parametr by definiuje kolumny, na podstawie ktÃ³rych chcemy sortowaÄ‡
nba.sort_values(by="Name")
```

```python
# moÅ¼emy rÃ³wnieÅ¼ sortowaÄ‡ w kolejnoÅ›ci descending
# domyÅ›lnie pandas umieszcza wartoÅ›ci NaN na koÅ„cu (niezaleÅ¼nie od kierunku sortowania)
nba.sort_values(by="Salary", ascending=False)
```

```python
# za pomocÄ… parametru na_position moÅ¼emy zdefiniowaÄ‡, czy wartoÅ›ci NaN majÄ… pojawiÄ‡ siÄ™ na poczÄ…tku/koÅ„cu
nba.sort_values(by="Salary", ascending=False, na_position="first")
```

## Sort a DataFrame with the sort_values Method II
- To sort by multiple columns, pass the `by` parameter a list of column names. Pandas will sort in the specified column order (first to last).
- Pass the `ascending` parameter a Boolean to sort all columns in a consistent order (all ascending or all descending).
- Pass `ascending` a list to customize the sort order *per* column. The `ascending` list length must match the `by` list.

```python
nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
# moÅ¼emy rÃ³wnieÅ¼ sortowaÄ‡ po kilku kolumnach, ale podajemy w parametrze listÄ™ kolumn
# jest to sortowanie sekwencyjne i wpisujemy kolejne kolumny zgodnie z sekwencjÄ… sortowania
# gÅ‚Ã³wne sortowanie jest na podstawie kolumny "Team", a nastÄ™pnie kolejne na podstawie kolumny "Name"
nba.sort_values(by=["Team", "Name"])
```

```python
# jeÅ¼eli chcemy sortowaÄ‡ rÃ³Å¼ne kolumny w rÃ³Å¼nych kolejnoÅ›ciach, to w parametrze ascending podajemy listÄ™ wartoÅ›ci True/False
nba.sort_values(by=["Team", "Name"], ascending=[True, False])
```

```python
# sortowanie jest wykonywane na kopii oryginalnego obiektu
nba.sort_values(by=["Position", "Salary"], ascending=[True, False])
```

```python
# jeÅ¼eli chcemy sortowanie zapisaÄ‡ w oryginalnym obiekcie, to musimy uÅ¼yÄ‡ przypisania
nba = nba.sort_values(by=["Position", "Salary"], ascending=[True, False])
nba
```

## Rank Values with the rank Method
- The `rank` method assigns a numeric ranking to each **Series** value.
- Pandas will assign the same rank to equal values and create a "gap" in the dataset for the ranks.

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba["Salary"]
```

```python
# jeÅ¼eli chcemy utworzyÄ‡ ranking na podstawie zdefiniowanych wartoÅ›ci, to moÅ¼emy wykorzystaÄ‡ metodÄ™ rank
# metoda rank zwrÃ³ci nam kolumnÄ™ z miejscem pracownika w rankingu
nba["Salary"].rank()
```

```python
# jeÅ¼eli wykorzystamy parametr ascending=False, to ranking zostanie zdefiniowany w kolejnoÅ›ci descending (od najwiÄ™kszych do najmniejszych)
# moÅ¼emy zwrÃ³ciÄ‡ te wartoÅ›ci jako liczby caÅ‚kowite
nba["Salary"].rank(ascending=False).astype("int")
```

```python
# ranking moÅ¼emy dodaÄ‡ jako kolejna kolumnÄ™
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
nba
```

```python
# jeÅ¼eli jakieÅ› rekordy z rankingu bÄ™dÄ… mieÄ‡ te same wartoÅ›ci, to powstaje remis i wartoÅ›Ä‡ rankingowe siÄ™ powtarzajÄ…
nba.sort_values(by="Salary", ascending=False).head(10)
```
